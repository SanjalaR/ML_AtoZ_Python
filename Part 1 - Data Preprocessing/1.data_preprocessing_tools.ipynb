# Data Preprocessing Tools

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
df = pd.read_csv('/content/Data.csv')
```

Matrix of features: Independent variables (columns with which the dependent variable will be predicted, contains info with which we can predict what we want to predict)


```
X = df.iloc[:, :-1].values  # all rows, all columns excluding last column
y = df.iloc[:, -1].values  # all rows, last column
```

```
X
```

```
y
```

## Taking care of missing data

Large dataset, 1% missing data -> delete record\
Else -> replace missing value with average/median/most frequent of column

```
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
```

Fit: connect imputer to matrix of features\
Transform: apply transformation by replacing missing value




```
imputer.fit(X[:, 1:3])  # all rows, columns 1 to 2 (exclude 3)
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```
print(X)
```

## Encoding categorical data

### Encoding the Independent Variable

One hot encoding -> turn categorical column into n diff columns (corr. to n categories). create binary vectors for each category

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
# transformer: what kind of transformation and for what columns. remainder: what to do with other columns
X = np.array(ct.fit_transform(X))
```

```
print(X)
```

### Encoding the Dependent Variable

Dependent variable -> Direct encoding is fine for binary outcome

```
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
```

```
print(y)
```

## Splitting the dataset into the Training set and Test set

Training set -> train model on existing observations\
Test set -> evaluate perf on new observations

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, train_size=0.8, random_state=1)
```

```
print(X_train)
```

```
print(X_test)
```

```
print(y_train)
```

```
print(y_test)
```

## Feature Scaling

Scaling all variables (features) so that values are in same scale. To avoid some features to be dominated by other features and not be considered by the ML model

Standardisation (*): x-mean(x)/standarddev(x) ->values around -3 to +3\
Normalisation: x-min(x)/max(x)-min(x) ->values bw 0 and 1

fit x train only. apply the same to transform x test

```
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()  # not applied to dummy variables from onehotencoding
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])  #leaving out the encoded columns, fit will calculate mean and standard deviation, transform will apply the formula
X_test[:, 3:] = sc.transform(X_test[:, 3:])
```
