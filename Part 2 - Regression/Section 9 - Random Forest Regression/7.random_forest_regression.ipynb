# Random Forest Regression

k random points from training set -> build decision tree for those k points -> build N number of such trees -> for new point, make each Ntree predict y, assign average

ensemble learning -> powerful -> change in one data point may affect 1/2 trees not whole forest

More adapted to highly dimensional datasets

## Importing the libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```
df = pd.read_csv('/content/Position_Salaries.csv')
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

## Training the Random Forest Regression model on the whole dataset

```
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(n_estimators=10, random_state=0)
reg.fit(X, y)
```

## Predicting a new result

```
reg.predict([[6.5]])
```

## Visualising the Random Forest Regression results (higher resolution)

```
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color='red')
plt.plot(X_grid, reg.predict(X_grid), color='blue')
plt.title('Truth or Bludd (RFR)')
plt.xlabel('Position level')
plt.ylabel('Salaries')
plt.show()
```

more trres, more splits, more steps

R squared -> <br>
r^2 = 1 - SSres/SStot<br>
SSres = sum of all (y-ypred)^2<br>
SStot = sum of all (y-yavg)^2<br>
SSres<SStot<br>
so r^2 is bw 0 and 1<br>
the better out model fits the data, the smaller SSres is, greater r^2.<br>
1 - sus<br>
0.9 - very good<br>
<0.7 - not great <br>
<0.4 - terrible <br>
<0 - no sense<br>
adding new x values -> SStot not change -> SSres decrease or remain same, never increase (Ordinary Least Squares method will minimise the result by finding b3 such that y-ypred is min, or make b3 as 0) -> model have lot of unnecessary variables that are increasing the r^2<br><br>
Adjusted r squared -><br>
adj r^2 = 1 - (1 - r^2)(n-1)/(n-k-1)<br>
k = no. of indep variables<br>
n = sample size<br>
k increases -> denominator decreases -> ratio increases -> adj r^2 decreases<br>
only add variables when they bring substantial improvement to model
